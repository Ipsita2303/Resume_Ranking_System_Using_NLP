```python
def rank_resumes(job_desc, resumes):
    job_embedding = get_embedding(job_desc)
    resume_embeddings = []
    for resume in resumes:
        resume_embeddings.append(get_embedding(resume))

    scores = cosine_similarity(
        [job_embedding],
        resume_embeddings)[0]
    return scores
```
